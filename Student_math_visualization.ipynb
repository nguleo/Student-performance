Data Visualization

```python
numerical_columns = data.select_dtypes(include=['int64', 'float64']).columns
# Display the correlation between the numerical columns
print(data[numerical_columns].corr())
```

```python
# 1. What factors contribute the most to student academic performance?
#We can explore correlations between the final grade (G3) and other factors like study time, parental education, support, or social activities.

plt.figure(figsize=(12, 6))
sns.heatmap(data[numerical_columns].corr(), annot=True)
plt.show()
```

```python
# Visualize the relationship between parental job status and math grades
plt.figure(figsize= (14,6))
# Mother's job
plt.subplot(1,2,1)
sns.boxplot(x='Mjob', y='G3', data=data)
plt.title("Mother's Job Status vs Math Grades")
plt.xlabel("Mother's Job Status")
plt.ylabel("Math Grade (G3)")
plt.xticks(rotation=45)

# Father's job
plt.subplot(1,2,2)
sns.boxplot(x='Fjob', y='G3', data=data)
plt.title("Father's Job Status vs Math Grades")
plt.xlabel("Father's Job Status")
plt.ylabel("Math Grade (G3)")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
#Factors Contributing the Most to Student Academic Performance

correlation_matrix = data[numerical_columns].corr()
g3_correlation = correlation_matrix['G3'].sort_values(ascending=False)

# Plotting the correlation heatmap for G3
plt.figure(figsize=(10,6))
sns.barplot(x=g3_correlation.index, y=g3_correlation.values)
plt.xticks(rotation=90)
plt.xlabel('Correlation with Final Grade (G3)')
plt.ylabel('Correlation Coefficient')
plt.show()
```

```python
# Effect of Parental Education Level on Student Performance
# Correlation heatmap for parental education (Medu, Fedu) and G3
plt.figure(figsize=(6,4))
sns.heatmap(data[['Medu', 'Fedu', 'G3']].corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation between Parental Education and Student Performance')
plt.show()

# Plot G3 vs Medu and Fedu
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.boxplot(x='Medu', y='G3', data=data)
plt.title('Mother\'s Education vs Final Grade (G3)')

plt.subplot(1,2,2)
sns.boxplot(x='Fedu', y='G3', data=data)
plt.title('Father\'s Education vs Final Grade (G3)')
plt.show()
```

```python
#Impact of School support and family support on G3

# Plot showing the impact of school support (schoolsup) and family support (famsup) on G3
plt.figure(figsize=(12,6))
sns.barplot(x='schoolsup', y='G3', hue='famsup', data=data)
plt.title('Impact of School Support (schoolsup) and Family Support (famsup) on Final Grade (G3)')
plt.ylabel('Average Final Grade (G3)')
plt.show()
```

```python
#Examine the relationship between family size and G3

plt.figure(figsize=(14, 6))

#Family size
sns.boxplot(x='famsize', y='G3', data=data)
plt.title("Family Size vs Math Grades")
plt.xlabel("Family Size")
plt.ylabel("Math Grade (G3)")

plt.show()
```

```python
# Visualize the relationship between study time and math grades
plt.figure(figsize=(10, 6))
sns.boxplot(x='studytime', y='G3', data=data)
plt.title('Study Time vs Math Grades')
plt.xlabel('Study Time')
plt.ylabel('Math Grade (G3)')
plt.show()
```

```python
# Reason they choose this school

plt.figure(figsize= (12,6))
sns.boxplot(x='reason', y='G3', data=data)
plt.title("Reason vs Math Grades")
plt.xlabel("Reason for Choosing School")
plt.ylabel("Math Grade (G3)")

plt.show()
```

```python
#Which sex has a better average performance
plt.figure(figsize= (12,6))
sns.barplot(x='sex', y='G3', data=data)
plt.title("Sex vs Math Grades")
plt.xlabel("Sex")
plt.ylabel("Math Grade (G3)")
```

```python
#What is the distribution of Math Grades
sns.displot(data=data, kind='hist', x='G3', kde='True')
```

```python
#What are the patterns of student absences and their impact on academic performance?

plt.figure(figsize=(14,7))
plt.title("Student Absences vs G3")
sns.barplot(x='G3', y='absences', data=data)
plt.xlabel('G3')
plt.ylabel('Absences')

plt.show()
```

```python
# Compare grades of students involved in extracurricular activities vs. not
sns.boxplot(x='activities', y='G3', data=data)
plt.title('Impact of Extracurricular Activities on Math Grades')
plt.show()
```

```python
# Workday alcohol consumption (Dalc) vs Math Grades
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(x='Dalc', y='G3', data=data)
plt.title('Workday Alcohol Consumption vs Math Grades')
plt.xlabel('Workday Alcohol Consumption (Dalc)')
plt.ylabel('Math Grade (G3)')

# Weekend alcohol consumption (Walc) vs Math Grades
plt.subplot(1, 2, 2)
sns.boxplot(x='Walc', y='G3', data=data)
plt.title('Weekend Alcohol Consumption vs Math Grades')
plt.xlabel('Weekend Alcohol Consumption (Walc)')
plt.ylabel('Math Grade (G3)')

plt.tight_layout()
plt.show()
```
